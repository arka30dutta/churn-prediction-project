```python
import pandas as pd
import numpy as np

# Read the CSV file into a DataFrame
df = pd.read_csv('variance.csv')

# Define the threshold for outlier detection (e.g., +/-3 sigma)
threshold = 3

# Define a function for outlier treatment using standardization
def treat_outliers(df):
    # Calculate the mean and standard deviation for each column
    mean = df.mean()
    std_dev = df.std()
    
    # Detect outliers using the +/-3 sigma approach
    lower_bound = mean - threshold * std_dev
    upper_bound = mean + threshold * std_dev
    
    # Replace outliers with the nearest bound
    df_clipped = df.clip(lower=lower_bound, upper=upper_bound, axis=1)
    
    return df_clipped

# Apply outlier treatment to each column
df_treated = treat_outliers(df)

# Write the treated DataFrame back to a CSV file
df_treated.to_csv('standardization.csv', index=False)
```

```python
# Read the CSV file into a DataFrame
df = pd.read_csv('variance.csv')

# Define the threshold for outlier detection (e.g., 1.5 times the IQR)
threshold = 1.5

# Iterate over each numerical column
for column in df.select_dtypes(include=['number']).columns:
    # Calculate the quartiles
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    
    # Calculate the Interquartile Range (IQR)
    IQR = Q3 - Q1
    
    # Define the lower and upper bounds
    lower_bound = Q1 - threshold * IQR
    upper_bound = Q3 + threshold * IQR
    
    # Identify outliers
    outliers = (df[column] < lower_bound) | (df[column] > upper_bound)
    
    # Replace outliers with NaN or the nearest bound
    #df[column] = df[column].mask(outliers, None)  # Replace with NaN
    # Alternatively, you can replace outliers with the nearest bound:
    df[column] = df[column].clip(lower=lower_bound, upper=upper_bound)
    
# Write the modified DataFrame back to a CSV file
df.to_csv('boxplot.csv', index=False)
```

```python
import pandas as pd

# Read the CSV file into a DataFrame
df = pd.read_csv('variance.csv')

# Define the threshold for outlier detection (e.g., 1.5 times the IQR)
threshold = 1.5

# Define a function for outlier treatment using capping and flooring
def treat_outliers(column):
    Q1 = column.quantile(0.25)
    Q3 = column.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - threshold * IQR
    upper_bound = Q3 + threshold * IQR
    # Cap and floor outliers
    column = column.clip(lower=lower_bound, upper=upper_bound)
    return column

# Apply outlier treatment to each column
for column in df.columns:
    df[column] = treat_outliers(df[column])

# Write the modified DataFrame back to a CSV file
df.to_csv('cf.csv', index=False)
```

```python
import pandas as pd

# Read the CSV file into a DataFrame
df = pd.read_csv('cf.csv')

# Define the threshold for correlation
threshold = 0.90  # You can adjust this threshold based on your requirement

# Calculate the correlation matrix
corr_matrix = df.corr().abs()

# Create a mask to identify highly correlated features
mask = (corr_matrix >= threshold) & (corr_matrix < 1)

# Get the names of highly correlated features
highly_correlated_features = set()
for col in mask.columns:
    correlated_cols = mask.index[mask[col]].tolist()
    if correlated_cols:
        highly_correlated_features.add(col)
        highly_correlated_features.update(correlated_cols)

# Remove highly correlated features from the DataFrame
df_filtered = df.drop(highly_correlated_features, axis=1)

# Write the filtered DataFrame back to a CSV file
df_filtered.to_csv('co.csv', index=False)
```

```python
import pandas as pd
import numpy as np

# Load your CSV dataset into a pandas DataFrame
data = pd.read_csv('co.csv')

# Drop any non-numeric columns if necessary
data = data.select_dtypes(include=[np.number])

# Calculate correlation matrix
correlation_matrix = data.corr()

# Calculate inverse correlation matrix
try:
    inv_correlation_matrix = np.linalg.inv(correlation_matrix.values)
except np.linalg.LinAlgError:
    print("Error: Matrix is singular. There may be perfect multicollinearity.")
    exit()

# Calculate VIF for each column
vif_values = np.diagonal(inv_correlation_matrix)

# Create a DataFrame to store column names and VIF values
vif_data = pd.DataFrame({
    'Column': data.columns,
    'VIF': vif_values
})

# Filter columns with VIF greater than 5
high_vif_columns = vif_data[vif_data['VIF'] > 5]['Column']

# Print or use the columns with high VIF
print("Columns with VIF greater than 5:")
print(high_vif_columns)
```
